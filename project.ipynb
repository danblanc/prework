#### What is the most expensive home of each population? (★★☆)

#### Normalize the variable of prices for each population and plot the 4 histograms in the same plot (★★★) 

```python
subset=sub[['price','level5']]

media = subset['price'].mean()
std_dev = subset['price'].std()

subset['price_normalized'] = (subset['price'] - media) / std_dev

subset['price_normalized']
```

```python
subset['level5'].unique()
```

```python
figure, axis = plt.subplots(2, 2)
  
subset_f = subset[subset['level5']=='Fuenlabrada']
subset_a = subset[subset['level5']=='Alcorcón']
subset_l = subset[subset['level5']=='Leganés']
subset_g = subset[subset['level5']=='Getafe']

# For Sine Function
axis[0, 0].hist(subset_f['price_normalized'])
axis[0, 0].set_title("Fuenlabrada")
  
# For Cosine Function
axis[0, 1].hist(subset_a['price_normalized'])
axis[0, 1].set_title("Alcorcón")
  
# For Tangent Function
axis[1, 0].hist(subset_l['price_normalized'])
axis[1, 0].set_title("Leganés")
  
# For Tanh Function
axis[1, 1].hist(subset_g['price_normalized'])
axis[1, 1].set_title("Getafe")
  
# Combine all the operations and display
plt.show()
```
